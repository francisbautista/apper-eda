1. Write a function to check whether a string is a palindrome

```python
def is_palindrome(str_in):
    letters = [x for x in str_in] # List letters in string
    letters.reverse() # Reverse list
    str_reversed = "".join(letters) # Join reversed list as a string
    if str_in == str_reversed:
        return True
    else:
        return False
```

```python
str_in = input('Input a string to check: ')
is_palindrome(str_in)
```

2. Write a function that checks whether a number is prime or not

```python
def is_prime(x):
    if x == 1: # 1 is neither prime nor composite
        return False
    if x == 2 or x == 3: # 2 and 3 are primes
        return True
    elif x % 2 == 0: # check if number divides 2 w/ zero remainder
        return False
    elif x % 3 == 0: # check if number divides 3 w/ zero remainder
        return False
    else:
        i = 5
        while i <= x ** 0.5: 
        # Check if number divides i w/ zero remainder only until square root of number
            if x % i == 0:
                return False
            else:
                i += 2
        return True
```

```python
int_in = int(input('Input a number to check if prime or not: '))
is_prime(int_in)
```

3. Create a function that takes in 3 parameters (integer n, integer a, integer b) and generates an nxn numpy matrix with random values from the range a to b inclusive

```python
def generate_matrix(n, a, b):
    rng = np.random.default_rng()
    return rng.integers(low=a, high=b, size=(n,n), endpoint=True, dtype=np.int64)
```

```python
n, a, b = [int(x) for x in input(
          'Input integers n,a,b separated by comma to generate an nxn numpy matrix w/ random values from a to b inclusive: ').split(',')]
generate_matrix(n,a,b)
```

5. Create a function that lists the first n prime numbers

```python
def list_first_n_primes(n):
    primes = [2]
    n_primes = len(primes)
    
    x = 3
    while n_primes < n:
        if is_prime(x):
            primes.append(x)
            n_primes = len(primes)
        x += 2
    
    print(f'First {n} primes: {primes}')
```

```python
n = int(input('Input a number of primes to list: '))
list_first_n_primes(n)
```

```python
idols = pd.read_csv('data/kpop_idols.csv')
idols
```

```python
idols.dtypes
```

```python
idols['Date of Birth'] = pd.to_datetime(idols['Date of Birth'])
idols.dtypes
```

- w/in the dataset, who is the oldest kpop idol?

```python
idols.sort_values(by=['Date of Birth'], axis=0).head()
```

W/in the dataset, the oldest kpop idol is **PSY**

- w/c group, using the mean of the members' ages combined, is the oldest?

```python
def compute_age(birth_date):
    today = pd.Timestamp.today()
    n_days = (today - birth_date).days
    return np.round_(n_days / 365, 2)
```

```python
idols['Age'] = idols['Date of Birth'].apply(compute_age)
idols
```

```python
idols.groupby('Group')['Age'].mean().sort_values(ascending=False)
```

Using the mean of the members' ages combined, the oldest group (excluding solo artists) is **Brown Eyed Girls** with a mean age of 38.9 years

```python
idols[idols['Group']=='Brown Eyed Girls']
```
