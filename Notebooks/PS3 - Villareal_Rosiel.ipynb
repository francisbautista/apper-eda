## Load data

I created a ``SpamHam_DataLoader()`` which does the ff:
1. ``read_files()``
    - reads the files in a directory, which has subdirectories "ham" and "spam"
2. ``load_data()``
    - takes multiple directories and stores ham and spam files in corresponding lists "hams" and "spams"
3. ``dissect_components()``
    - dissects an email into its components "subject" and "body" and replaces email addresses in the body with marker "email_address"
    - I got the idea to use markers like "emailaddress" to replace email addresses from this __[github blog](https://sdsawtelle.github.io/blog/output/spam-classification-part1-text-processing.html)__.
4. ``create_df()``
    - structures the email data into 1 dataframe with columns "status", "subject", "body"

References on creating classes and methods:
- __[Python classes](https://docs.python.org/3/tutorial/classes.html)__
- __[Python getattr()](https://docs.python.org/3/library/functions.html#getattr)__

```python
class SpamHam_DataLoader():


    def __init__(self):
        self.hams = []
        self.spams = []


    def read_files(self, data_dir):
        """
        Takes a directory, finds subdirectories 'ham' and 'spam',
        And appends files in the subdirectories to a ham list and a spam list

        Args:
            - rootdir (str): path to directory where ham and spam texts are located

        Returns:
            - ham_list (list): a list of texts from ham emails found in the directory
            - spam_list (list): a list of texts from spam emails found in the directory
        """
        ham_list = []
        spam_list = []

        for directories, subdirs, files in os.walk(data_dir):
            if (os.path.split(directories)[1]  == 'ham'):
                for filename in files:      
                    with open(os.path.join(directories, filename), encoding="latin-1") as f:
                        data = f.read()
                        ham_list.append(data)

            if (os.path.split(directories)[1]  == 'spam'):
                for filename in files:
                    with open(os.path.join(directories, filename), encoding="latin-1") as f:
                        data = f.read()
                        spam_list.append(data)

        return ham_list, spam_list


    def load_data(self, data_dir_list):
        """
        Takes a list of directories, each with subdirectories 'ham' and 'spam',
        Reads the files inside the subdirectories,
        And appends them to the ham and spam lists

        Args:
        - data_dir_list (list): a list of paths (str) to check for files
        """
        for data_dir in data_dir_list:
            ham_list, spam_list = self.read_files(data_dir)
            self.hams.extend(ham_list)
            self.spams.extend(spam_list)


    def dissect_components(self, text):
        """
        Dissects a text into its components 'subject' and 'body',
        Replaces email addresses in the body with marker <email_address>

        Args:
            - text (str): email containing a subject, body, and other header info

        Returns:
            - subject (str): email subject (or subjects if forwarded email)
            - body (str): email message body
        """
        subj_regx = re.compile(r'[s|S]ubject\s*:(.+)')
        subjects = subj_regx.findall(text)
        subject = ' '.join(set(subjects))

        senders = re.compile(r'from\s*:(.+)')
        forwarders = re.compile(r'forwarded by\s*(.+)')
        recipients = re.compile(r'to\s:(.+)')
        cc = re.compile(r'cc\s:(.+)')

        emailadd = re.compile(r'(\w+\s*@\s*\w+\s*\.\s*\w+)')

        # Replace every name/email address with a marker
        for comp in [senders, forwarders, recipients, cc, emailadd]:
            text = re.sub(comp, '<email_address>', text)

        # Remove subjects from email body
        body = re.sub(subj_regx, '', text)

        return subject, body


    def create_df(self):
        """
        Takes the lists of spam and ham texts,
        Dissects each text into subject and body,
        And creates a dataframe containing dissected texts

        Returns:
            - df (dataframe): A dataframe of ham and spam emails with 'subject', 'body', and 'status' as columns
        """
        iterator = iter(dl.spams)
        done, subjects, bodies = False, [], []
        while not done:
            try:
                subject, body = dl.dissect_components(next(iterator))
                subjects.append(subject)
                bodies.append(body)
            except StopIteration:
                done = True
        spam_df = pd.DataFrame({'subject': subjects, 'body': bodies, 'status': 'spam'})

        iterator = iter(dl.hams)
        done, subjects, bodies = False, [], []
        while not done:
            try:
                subject, body = dl.dissect_components(next(iterator))
                subjects.append(subject)
                bodies.append(body)
            except StopIteration:
                done = True
        ham_df = pd.DataFrame({'subject': subjects, 'body': bodies, 'status': 'ham'})

        df = pd.concat([spam_df, ham_df], ignore_index=True)

        return df
```

# Preprocess data

## Clean messages by removing urls, punctuation, numbers, and extra whitespaces
## Lemmatize words

I created a ``Preprocessor()`` which does the ff:
1. ``remove_urls()``
    - replaces URLs, i.e., website links that begin with "http", or "https", with marker "url"
2. ``remove_punctuation()``
    - punctuation marks using the list of punctuation marks from the string library
3. ``remove_numbers()``
    - numbers since they could be phone numbers, datetime, money, or just random nonsense; hard to determine which because there are too many formats to write phone numbers, datetime, or money
4. ``remove_extra_whitespaces()``
    - extra whitespaces & newline markers
5. ``lemmatize_words()``
    - lemmatizes, or reduces words to their root forms while ensuring that the root forms are still valid words

I used the __[re library](https://docs.python.org/3/library/re.html)__ and this __[regex cheatsheet](https://www.w3schools.com/python/python_regex.asp)__ to find and replace urls, punctuation marks, numbers, and extra whitespaces.

```python
string.punctuation
```

```python
class Preprocessor():


    def remove_urls(self, text):
        """
        Replaces URLS that begin with 'http' or 'https' in a text with marker <url>
        """
        http = re.compile(r'http\s*.+')
        https = re.compile(r'http\s*.+')

        # Replace every url with a marker
        for link in [http, https]:
            text = re.sub(link, '<url>', text)

        return text


    def remove_punctuation(self, text):
        """
        Replaces punctuation marks in a text with whitespaces
        """
        puncts = list(string.punctuation)
        for punct in puncts:
            text = text.replace(punct, '')
        return text


    def remove_numbers(self, text):
        """
        Replaces numbers in a text with whitespaces
        """
        return re.sub('\d+', '', text)


    def remove_extra_whitespaces(self, text):
        """
        Removes extra whitespaces in a text
        """
        text = re.sub('\s+', ' ', text) # retain a single whitespace only
        text = text.strip(' ') # strip of leading & trailing whitespaces
        return text

    def lemmatize_words(self, text):
        """
        Lemmatizes words in a text
        """
        lemmatizer = WordNetLemmatizer()
        return ' '.join([lemmatizer.lemmatize(word) for word in text.split()])
```

```python
def preprocess_df(df, col_to_preprocess):
    """
    Preprocesses a dataframe by removing URLS, punctuation marks, numbers, and extra whitespaces sequentially;
    Creates new columns in the dataframe for intermediate outputs;
    And returns the preprocessed dataframe

    Args:
        - df (dataframe): dataframe containing column values to preprocess 
        - col_to_preprocess (str): name of column to preprocess
    """
    preproc = Preprocessor()
    preprocess_funcs = ['remove_urls', 'remove_punctuation',
                        'remove_numbers', 'remove_extra_whitespaces',
                        'lemmatize_words']
    col_prefix = col_to_preprocess
    # Apply preprocessing steps in a sequential manner
    for idx, func in enumerate(preprocess_funcs):
        col_name = col_prefix+'_'+preprocess_funcs[idx]

        if idx != 0: # preprocess the output of the previous preprocessing step
            col_to_preprocess = col_prefix+'_'+preprocess_funcs[idx-1]

        df[col_name] = df.apply(lambda row:
                                getattr(preproc, preprocess_funcs[idx])(row[col_to_preprocess]),
                                axis=1)

    return df
```

## Vectorize words using tf-idf

I removed English stopwords so that only meaningful words are left. I also used the parameter ``min_df`` in ``TfidfVectorizer()`` to remove very rare words and set the minimum word frequency for inclusion in the tf-idf matrix.
> ``min_df``: When building the vocabulary ignore terms that have a document frequency strictly lower than the given threshold. This value is also called cut-off in the literature. If float in range of [0.0, 1.0], the parameter represents a proportion of documents, integer absolute counts. This parameter is ignored if vocabulary is not None.

I built a pipeline that includes the tf-idf vectorization and classification steps. I also got the intermediate output from the tf-idf vectorization step, i.e., the resulting tf-idf matrix, using the snippet of code from __[stackoverflow: How to return a dataframe from sklearn TFIDF Vectorizer within pipeline?](https://stackoverflow.com/questions/52759118/how-to-return-a-dataframe-from-sklearn-tfidf-vectorizer-within-pipeline)__.

```python
class DenseTfidfVectorizer(TfidfVectorizer):

    def transform(self, raw_documents, copy=True):
        X = super().transform(raw_documents, copy=copy)
        df = pd.DataFrame(X.toarray(), columns=self.get_feature_names())
        return df

    def fit_transform(self, raw_documents, y=None):
        X = super().fit_transform(raw_documents, y=y)
        df = pd.DataFrame(X.toarray(), columns=self.get_feature_names())
        return df
```

## Fit a Naive Bayes classifier on the training set

> The multinomial Naive Bayes classifier is suitable for classification with discrete features (e.g., word counts for text classification). The multinomial distribution normally requires integer feature counts. However, in practice, fractional counts such as tf-idf may also work.
- Source: __[Multinomial NB](https://scikit-learn.org/stable/modules/generated/sklearn.naive_bayes.MultinomialNB.html)__

```python
def train_model(train_x, train_y, stop_words = 'english', min_df = 0.01):
    """
    Transforms the train set using tf-idf vectorization,
    Classifies the data using Multinomial Naive Bayes,
    Shows the classification report with metrics and the resulting tf-idf matrix
    
    Args:
        - train_x (pandas series): training set of email message bodies
        - train_y (pandas series): training set statuses (either "ham" or "spam")
        - stop_words: 'english' (str) or None
        - min_df (float): from 0 to 1, min frequency of a term to include it in tf-idf matrix

    Returns:
        - pipe (sklearn.pipeline.Pipeline): Pipeline object that can be used to predict(), predict_proba()
    """
    print(f'Transforming the train set using tf-idf vectorization ' +
          f'using stop_words: {stop_words} and min_df {min_df}')
    pipe = Pipeline([('tfidf_vectorizer', DenseTfidfVectorizer(stop_words = stop_words, min_df = min_df)),
                     ('nb_classifier', MultinomialNB())])
    pipe.fit(train_x, train_y)
    
    print(f'Classifying emails using Multinomial Naive Bayes...')
    y_pred = pipe.predict(train_x) # Predict on training set
    print('Classification report:')
    print(metrics.classification_report(train_y, y_pred)) # View evaluation metrics on whole training set

    print(f'Showing output of tf-idf vectorization...')
    tfidf_matrix = pipe.steps[pipe.steps[0]=='tfidf_vectorizer'][1].fit_transform(train_x)
    print(f'Tf-idf vectorization produced {tfidf_matrix.shape[1]} features for {tfidf_matrix.shape[0]} documents')
    print(tfidf_matrix)
    return pipe
```

```python
pipe = train_model(train_x, train_y, stop_words = 'english', min_df = 0.01)
```

## Compute classifier metrics on the validation set

Cross-validation can be done to evaluate different classification metrics (accuracy, precision, recall, f1-score).

```python
# Create scorers to use in cross_validate()


def macro_precision_score(y_true,y_pred):
    macro_precision = metrics.precision_score(y_true, y_pred, average='macro')
    return macro_precision


def macro_recall_score(y_true,y_pred):
    macro_recall = metrics.recall_score(y_true, y_pred, average='macro')
    return macro_recall


def macro_f1_score(y_true,y_pred):
    macro_f1= metrics.f1_score(y_true, y_pred, average='macro')
    return macro_f1


# Run cross-validation, computing the mean and std across folds for different metrics
def compute_cv_scores(clf, X, y, cv=5):
    macro_prec_scorer = metrics.make_scorer(macro_precision_score)
    macro_recall_scorer = metrics.make_scorer(macro_recall_score)
    macro_f1_scorer = metrics.make_scorer(macro_f1_score)
    scores = cross_validate(clf, X, y, cv=cv,
                            scoring={'accuracy': 'accuracy',
                                    'precision': macro_prec_scorer,
                                    'recall': macro_recall_scorer,
                                    'f1': macro_f1_scorer}
                           )
    print(f"Accuracy: {np.round_(scores['test_accuracy'].mean() *100, 2)} +/- "+
          f"{np.round_(scores['test_accuracy'].std() *100, 2)} %")
    print(f"Precision: {np.round_(scores['test_precision'].mean() *100, 2)} +/- "+
          f"{np.round_(scores['test_precision'].std() *100, 2)} %")
    print(f"Recall: {np.round_(scores['test_recall'].mean() *100, 2)} +/- "+
          f"{np.round_(scores['test_recall'].std() *100, 2)} %")
    print(f"F1-score: {np.round_(scores['test_f1'].mean() *100, 2)} +/- "+
          f"{np.round_(scores['test_f1'].std() *100, 2)} % \n")


def validate_model(model, valid_x, valid_y, cv=5):
    y_pred = model.predict(valid_x) # Predict on validation set
    print('Classification report when predicting on whole set:')
    print(metrics.classification_report(valid_y, y_pred)) # View evaluation metrics on whole validation set

    print('Cross-validation scores: \n')
    compute_cv_scores(model, valid_x, valid_y, cv=cv) # Run cross-validation

    print('Confusion matrix when predicting on whole set:')
    metrics.plot_confusion_matrix(model, valid_x, valid_y) # Visualize confusion matrix
```

## Perform error analysis on validation set

I got the idea of running error analysis on a validation set to gain insights on cases that the model misclassified and creating new features from this __[nltk chapter on classifying text](https://www.nltk.org/book/ch06.html)__.

```python
def run_error_analysis(model, valid_x, valid_y, n_cases):
    # Predict probability of email being ham or spam
    probs =  model.predict_proba(valid_x)
    y_pred = model.predict(valid_x)

    # Create dataframe with predictions and probabilities
    valid_df = pd.DataFrame({'body': valid_x,
                         'actual_status': valid_y,
                         'predicted_status': y_pred.T,
                         'prob_ham': probs[:,0],
                         'prob_spam': probs[:,1]})

    # Create a dataframe for mistakes, or cases the model misclassified
    mistakes = valid_df.loc[valid_df['actual_status']!=valid_df['predicted_status']]

    # Look at n_cases SPAMS misclassified as hams with highest probabilities
    misleading_spam = mistakes.sort_values(by=['prob_ham'], ascending=False).head(n_cases)['body'].values

    print('Spam emails misclassified as ham: \n')
    for i in range(n_cases):
        print(f'Case {i}:\n')
        print(misleading_spam[i])
        print('\n')

    # Look at n_cases HAMS misclassified as spams with highest probabilities
    misleading_ham = mistakes.sort_values(by=['prob_spam'], ascending=False).head(n_cases)['body'].values

    print('Ham emails misclassified as spam: \n')
    for i in range(n_cases):
        print(f'Case {i}:\n')
        print(misleading_ham[i])
        print('\n')
```

### Add a feature for counting unusual words

- An insight I got from running error analysis, specifically, on Case 0 for Spam emails misclassified as spam, is that spam emails containing many unusual words can be misclassified as ham because the model ignores those unusual words. In other words, even if those unusual words could just be random nonsense, they do not affect how the model classifies the email.
- Thus, maybe one solution I could try to address the concern above is to add a feature that counts unusual words.
- I read this __[nltk chapter on wordlist corpora](https://www.nltk.org/book/ch02.html)__ and used the function ``unusual_words()`` in Section 4.1 Wordlist Corpora to find unusual or misspelt words in the message bodies using nltk's Wordlist corpus.

```python
print(f'The nltk Wordlist corpus contains {len(words.words())} English words.')
```

Some notes:
- I just want to check how much time it takes to count the number of unusual words for a given number of emails, so I'll use the time library here.
- Since the feature for the number of unusual words consists of integer counts, it would be very large compared to the tf-idf values. I applied log transformation to reduce its magnitude and somehow normalize it.
- But because some emails may have 0 unusual words, getting the log transform will result to inf. To remedy this, I just added 1 to ``num_unusual_words`` for all observations so that there are no 0 values and then applied log transformation again. I already added 1 in ``count_unusual_words()``.

```python
def count_unusual_words(text):
    text_vocab = set(w.lower() for w in text.split() if w.isalpha())
    english_vocab = set(w.lower() for w in words.words())
    unusual = text_vocab - english_vocab
    return len(sorted(unusual)) + 1
```

```python
def add_feature_unusual(df, col_to_process):
    print(f'Counting the number of unusual words in text in feature "{col_to_process}"')
    start_time = time.time()
    df['num_unusual_words'] = df.apply(lambda row: count_unusual_words(row[col_to_process]), axis=1)
    print(f'It took {time.time() - start_time} seconds to '+
          f'count the number of unusual words for {len(df)} emails.')

    print('Applying log transformation on the number of unusual words...')
    df['log_num_unusual_words'] = np.log(df['num_unusual_words'])

    return df
```

# Interpret results

I got the idea to use odds ratio to interpret model results in a concrete way from this __[nltk chapter on classifying text](https://www.nltk.org/book/ch06.html)__.
The function ``create_feature_prob()`` does the ff:
- For each term in the corpora, get its empirical log probability given a class, $P(term|class)$. 
- Compute the probability to see each term given that the email is ham or spam.
- Compute the odds ratio for spam:ham, i.e., ratio of seeing the term in spam emails to seeing the term in ham emails. For example, if the word *prescription* has an odds ratio of 73, it means that emails with the word *prescription* are **spam** 73 times more often than they are ham.
- Note that the computed probabilities and odds ratios are not the actual ones because I used tf-idf weights instead of integer counts. They're more of a proxy.

```python
print(pipe.steps)
```

```python
def create_feature_prob(vectorizer, classifier):
    # Compute empirical log probability of features given a class, P(x_i|y).
    feature_log_prob = pd.DataFrame(list(zip(vectorizer.get_feature_names(), classifier.feature_log_prob_[0].T,
                                             classifier.feature_log_prob_[1].T)),
                                    columns=['term', 'log_prob_ham', 'log_prob_spam'])

    # Compute the probability to see each term given that the email is ham or spam
    feature_log_prob['prob_ham'] = feature_log_prob.apply(lambda row: np.exp(row['log_prob_ham']), axis=1)
    feature_log_prob['prob_spam'] = feature_log_prob.apply(lambda row: np.exp(row['log_prob_spam']), axis=1)

    # Compute odds ratio of seeing term in spam vs. seeing it in ham
    feature_log_prob['OR_spam_ham'] = feature_log_prob['prob_spam']/feature_log_prob['prob_ham']
    return feature_log_prob

# I used the first model I fit, i.e., without the extra feature on the number of unusual words
# because adding that feature didn't improve the performance
feature_log_prob = create_feature_prob(pipe.steps[0][1], pipe.steps[1][1])
feature_log_prob
```

```python
def identify_most_common(feature_log_prob, sort_by, n_terms):
#     feature_log_prob.sort_values(sort_by, ascending=False).head(n_terms)
    
    plt.figure(figsize=(10,5))
    ax = sns.barplot(x='term',
            y=f'{sort_by}', data=feature_log_prob,
            order=feature_log_prob.sort_values(sort_by, ascending=False).head(n_terms).term)
    ax.tick_params(axis='x', rotation=90)
    ax.set_title(f'Terms with highest {sort_by}')
```

**Results interpretation**
- The most common terms found in ham emails are ``emailaddress``, the marker I used to replace the names and email addresses of senders and recipients, and ``enron``. 
    - Maybe this is due to spam emails in the dataset not having the sender and recipient information. In reality, the senders' email addresses can also be obtained from spam emails.
    - But the email address can still be used as a feature. For example, if person A has had previous email correspondence with person B in the past, then future emails coming from person B's email address are likely not spam. Person A most likely has regular email correspondence with specific people they work with, so emails coming from that list of specific people can be tagged as not spam. 
    - Aside from having a list of email contacts, it's also possible to check the email domain, or what comes after ``@``, to see if the email came from a company account. For example, for GCash employees, emails coming from addresses that end with ``@gcash.com`` can be tagged as not spam. 
- Meanwhile, the most common terms found in spam emails are ``url``, the marker I used to replace web addresses beginning with either ``http`` or ``https``, and ``software``. Indeed one telling sign that an email is spam is if it invites users to click suspicious links and visit unsafe websites. Another sign is if the email asks users to download ``software`` that might contain malware. 
- Thus, spam classification models can be created based not only on the content of the emails (e.g. if email has specific words), but also on the subjects of the emails and email addresses of the senders.

---
**Limitations**
- By running error analysis and looking at some cases that the model misclassified, I found out that some of the emails were written in non-English languages. Because most of the emails in the dataset were in English, the resulting tf-idf matrix consisted mostly of English words. Thus, the model found it difficult to classify non-English emails that contained many words that the model hasn't seen before.  
- Some spam emails contained nonsense or random permutations of letters. To detect that nonsense, I tried to add a feature based on the number of unusual words in the email, where an unusual word is something that's not in nltk's corpus of 236,736 English words. However, because of the multiple language problem, non-English emails would have high values for that feature. It would be possible to use corpora of multiple languages, but implementation would be really slow, especially for large datasets.
- Other spam emails contained valid English words that when put together make no sense. For example, one spam email had the words ``strong pasteboard anonymous square sprite capital`` in that specific sequence. When the words are taken individually, they don't look like words common in spam, but when they appear in one sentence, they look like garbage. The takeaway from this example is that the context of words are used in sentences matters. Maybe for future implementations, the occurrence of multiple words together in a text could be used as features.
